```python
import pandas as pd
import requests
from termcolor import colored
```

```python
files = pd.read_excel("CIT_Newly_added_Catalog_0521.xlsx")
files
```

```python
for i in range(1, len(files)):
    if (files["Url"][i].endswith(".pdf") or files["Url"][i].endswith(".pdf#")):
        r = requests.get(files["Url"][i], allow_redirects=True)                
        with open("CIT_Plans/ "+files["Plan Name"][i][0:232], 'wb') as plan:
            plan.write(r.content)
        print("Plan "+str(i)+" "+ files["Plan Name"][i][0:232]+" is downloaded successfully.")
    else:
        print(colored("Plan "+str(i)+" "+ files["Plan Name"][i][0:232]+" doesn't have accessible links.", 'red'))
    
```
